```python
cols = dataSet.columns
freq = np.load(define_.FREQ_ITEMS_G1_TRAIN_FILE_PATH,allow_pickle=True)
df = functions.freqItemToDF(freq, cols)
supp = np.load(define_.SUPP_DATA_FILE_PATH,allow_pickle=True)
df.head()
# len(df)
```

```python
rules= functions.generateRules(freq.tolist(),supp.tolist(), minConf=0.1)
rules
```

```python
cols = [define_.ITEM_A, define_.ITEM_B,define_.CONFIDENCE_AB,define_.CONFIDENCE_BA, define_.SUPPORT_AB,define_.SUPPORT_A, define_.SUPPORT_B,define_.LIFT]
final_Rules = pd.DataFrame(rules, columns=cols)
final_Rules[define_.ITEM_A] = final_Rules.apply(lambda row: functions.convertToStringList(str(row[define_.ITEM_A])), axis=1)
final_Rules[define_.ITEM_B] = final_Rules.apply(lambda row: functions.convertToStringList(str(row[define_.ITEM_B])), axis=1)
final_Rules.to_csv(define_.FINAL_RULES_G1_TRAIN_FILE_PATH)
final_Rules
```
